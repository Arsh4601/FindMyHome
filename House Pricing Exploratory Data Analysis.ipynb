# Handeling Missing Values

```python
data1=data.copy()

data1
```

```python
#Missing values:  LotFrontage, Alley,PoolQC,Fence,Miscfeature.

data1["LotFrontage"].fillna(0.0,inplace=True)
data1["Alley"].fillna("No Alley",inplace=True)
data1["PoolQC"].fillna("No Pool",inplace=True)
data1["Fence"].fillna("No Fence",inplace=True)
data1["MiscFeature"].fillna("No MiscFeature",inplace=True)
```
